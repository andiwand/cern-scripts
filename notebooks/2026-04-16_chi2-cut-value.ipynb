```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

```python
def powspace(start, stop, power, num):
    start = np.power(start, 1/float(power))
    stop = np.power(stop, 1/float(power))
    return np.power( np.linspace(start, stop, num=num), power)

p_values = powspace(0, 0.5, 10, 100)
gauss_values = scipy.stats.norm.ppf(1-p_values)
gauss_sq_values = gauss_values**2
chi2_df1_values = scipy.stats.chi2.ppf(1-p_values, df=1)
chi2_df2_values = scipy.stats.chi2.ppf(1-p_values, df=2)
chi2_df3_values = scipy.stats.chi2.ppf(1-p_values, df=3)
```

```python
plt.gca().set_xlabel("1-pvalue")
plt.gca().set_ylabel("cut")

plt.gca().set_xscale("log")

plt.plot(p_values, gauss_values, label="N")
plt.plot(p_values, gauss_sq_values, label="$N^2$")
plt.plot(p_values, chi2_df1_values, label=r"$\chi^2_{df=1}$")
plt.plot(p_values, chi2_df2_values, label=r"$\chi^2_{df=2}$")
plt.plot(p_values, chi2_df3_values, label=r"$\chi^2_{df=3}$")

plt.legend();
```

```python
plt.gca().set_xlabel("1-pvalue")
plt.gca().set_ylabel("cut")

plt.gca().set_xscale("log")

mask = (p_values > 1e-15) & (p_values < 1e-3)

plt.plot(p_values[mask], gauss_sq_values[mask]/gauss_sq_values[mask], label="N")
plt.plot(p_values[mask], chi2_df1_values[mask]/gauss_sq_values[mask], label=r"$\chi^2_{df=1}$")
plt.plot(p_values[mask], chi2_df2_values[mask]/gauss_sq_values[mask], label=r"$\chi^2_{df=2}$")
plt.plot(p_values[mask], chi2_df3_values[mask]/gauss_sq_values[mask], label=r"$\chi^2_{df=3}$")

plt.legend();
```

```python
for c in [9, 15, 25, 50, 100]:
    print(f"cut {c}")
    print(f"  norm2     -> 1-p = {1-scipy.stats.norm.cdf(c**0.5)}")
    print(f"  chi2 df=1 -> 1-p = {1-scipy.stats.chi2.cdf(c, df=1)}")
    print(f"  chi2 df=2 -> 1-p = {1-scipy.stats.chi2.cdf(c, df=2)}")
    print(f"  chi2 df=3 -> 1-p = {1-scipy.stats.chi2.cdf(c, df=3)}")
```
