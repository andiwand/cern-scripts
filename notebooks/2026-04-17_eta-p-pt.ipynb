```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
eta_values = np.linspace(0, 4, 100)
theta_values = 2*np.arctan(np.exp(-eta_values))
p_scale_values = 1/np.sin(theta_values)
```

```python
plt.gca().set_xlabel(r"$\eta$")
plt.gca().set_ylabel(r"p scale")

plt.plot(eta_values, p_scale_values);
```

```python
plt.gca().set_yscale("log")

plt.gca().set_xlabel(r"$\eta$")
plt.gca().set_ylabel(r"p scale")

plt.plot(eta_values, p_scale_values);
```
