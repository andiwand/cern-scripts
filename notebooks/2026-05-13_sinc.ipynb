```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(-100, -1, 10000)
xs = 10**xs

ys_np = np.sin(xs) / xs

def sinc(x):
    x = np.array(x)
    r = np.zeros(x.shape)
    mask_small = np.abs(x) < 1e-6
    r[mask_small] = 1 - x[mask_small]**2 / 6
    r[~mask_small] = np.sin(xs[~mask_small]) / xs[~mask_small]
    return r

ys_c = sinc(xs)
```

```python
plt.plot(xs, ys_np)
plt.plot(xs, ys_c)

plt.xscale("log")
plt.yscale("log")
```

```python
plt.hlines(1, -1, 1, color="C0")
plt.plot(xs, ys_c / ys_np, color="C1")

plt.xscale("log")
plt.yscale("log")
```

```python
ys_sin = np.sin(xs)

plt.plot(xs, ys_sin)

plt.xscale("log")
plt.yscale("log")
```
